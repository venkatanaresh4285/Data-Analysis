```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('googleplaystore.csv')
Imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
Imputer.fit(df.iloc[: ,2:3].values)
df=df.dropna()
```

# How many free apps are there in Art and Design

```python
df=df.values
```

```python
df[0]
```

```python
c=0
for i in df:
    if(i[1]=='ART_AND_DESIGN' and i[6]=='Free'):
        c+=1
print("There are ",c,"Number of art and design free apps")
```

# How many paid apps are there in Art and Design

```python
c=0

for i in df:
    if(i[1]=='ART_AND_DESIGN' and i[6]=='Paid'):
        c+=1

print("There are ",c,"Number of paid apps in ART AND DESIGN")
    
```

# How many apps are there rating greater than 4.5

```python
c=0
for i in df:
    if(i[1]=='ART_AND_DESIGN' and i[2]>4.5):
        c+=1       
print("There are ",c,"apps are greater than 4.5")
```

# How many Apps are free and greater than 4.5

```python
c=0
for i in df:
    if(i[1]=='ART_AND_DESIGN' and i[2]>4.5 and i[6]=='Free'):
        c+=1
print("There are ",c ,"Number of free appps which rating greater than 4.5")
```
