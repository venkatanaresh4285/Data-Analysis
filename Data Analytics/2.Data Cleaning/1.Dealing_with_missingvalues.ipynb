## Filling the missing values-imputation
   1.Numerical-mean
 

```python
data3=data1.copy()
```

```python
data3
```

```python
data3.head()
```

```python
mean_value=data3["reviews_per_month"].mean()
data3["reviews_per_month"].fillna(mean_value,inplace=True)
```

```python
data3.head()
```

```python
data3.isna().sum()
```

## Filling the Categorial data

```python
data3.nunique()
```

```python
data3["reviews_per_month"].fillna(data3["reviews_per_month"].value_counts().index[0],inplace=True)
```

```python
data3.isna().sum()
```

```python
data4=data.copy()
```

```python
data4.isna().sum()
```
